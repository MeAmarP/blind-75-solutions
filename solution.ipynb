## Problem 1: Two Sum

Category: Array

Difficulty: Easy

### Problem Description

Given an integer array `nums` and an integer `target`, return the indices of the two numbers such that they add up to `target`.

You may assume exactly one valid answer exists, and the same element cannot be used twice.

### Expected Input

- `nums`: list of integers
- `target`: integer

### Expected Output

- A list of two indices

### Solution 1: Brute Force Approach

Check every possible pair until the pair sum matches the target.

#### Algorithm

1. Loop through the array with index `i`.
2. For each `i`, loop through the remaining indices `j`.
3. Check whether `nums[i] + nums[j] == target`.
4. If yes, return `[i, j]`.
5. If no pair is found, return an empty list.


```python
def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use a hash map to remember previously seen values and their indices.

#### Algorithm

1. Create an empty dictionary `seen`.
2. Traverse the array once.
3. For each number, compute `target - num`.
4. If the complement is already in `seen`, return the stored index and current index.
5. Otherwise store the current number and index.
6. Return an empty list if no solution is found.


```python
def two_sum_optimal(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i

    return []
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(n)`


## Problem 2: Best Time to Buy and Sell Stock

Category: Array

Difficulty: Easy

LeetCode: [Link](https://leetcode.com/problems/best-time-to-buy-and-sell-stock/)

### Problem Description

Given an array `prices` where `prices[i]` is the price of a stock on day `i`, find the maximum profit you can achieve by choosing one day to buy and a later day to sell.

If no profit is possible, return `0`.

### Expected Input

- `prices`: a list of integers where each value represents the stock price on a given day

### Expected Output

- An integer representing the maximum profit from one buy and one sell

### Solution 1: Brute Force Approach

Try every possible buy day and every possible sell day after it, then keep track of the largest profit.

#### Algorithm

1. Initialize `max_profit` as `0`.
2. Loop through each day `i` as the buy day.
3. For each `i`, loop through every later day `j` as the sell day.
4. Compute `prices[j] - prices[i]`.
5. Update `max_profit` if the computed profit is larger.
6. Return `max_profit` after checking all pairs.


```python
def max_profit_bruteforce(prices):
    max_profit = 0

    for i in range(len(prices)):
        for j in range(i + 1, len(prices)):
            profit = prices[j] - prices[i]
            max_profit = max(max_profit, profit)

    return max_profit
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Track the minimum price seen so far while scanning the array once. At each day, treat the current price as a possible selling price and compute the profit against the cheapest earlier buy price.

#### Algorithm

1. Initialize `min_price` as the first price.
2. Initialize `max_profit` as `0`.
3. Traverse the array from left to right.
4. Update `min_price` whenever a smaller price is found.
5. Compute the profit from selling today using `current_price - min_price`.
6. Update `max_profit` if that profit is larger.
7. Return `max_profit`.


```python
def max_profit_optimal(prices):
    min_price = float("inf")
    max_profit = 0

    for price in prices:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)

    return max_profit
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`


## Problem 3: Contains Duplicate

Category: Array

Difficulty: Easy

LeetCode: [Link](https://leetcode.com/problems/contains-duplicate/)

### Problem Description

Given an integer array `nums`, return `True` if any value appears at least twice in the array. Return `False` if every element is distinct.

### Expected Input

- `nums`: a list of integers

### Expected Output

- A boolean value indicating whether any duplicate exists

### Solution 1: Brute Force Approach

Compare every element with every later element and check whether any pair has the same value.

#### Algorithm

1. Loop through each index `i`.
2. For each `i`, loop through all later indices `j`.
3. Compare `nums[i]` and `nums[j]`.
4. If they are equal, return `True`.
5. If all pairs are checked and no duplicate is found, return `False`.


```python
def contains_duplicate_bruteforce(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True

    return False
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use a set to track seen values. If a number is already in the set, a duplicate exists.

#### Algorithm

1. Create an empty set called `seen`.
2. Traverse the array once.
3. For each number, check whether it is already in `seen`.
4. If yes, return `True`.
5. Otherwise add it to `seen`.
6. Return `False` if the loop finishes without finding a duplicate.


```python
def contains_duplicate_optimal(nums):
    seen = set()

    for num in nums:
        if num in seen:
            return True
        seen.add(num)

    return False
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(n)`


## Problem 4: Product of Array Except Self

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/product-of-array-except-self/)

### Problem Description

Given an integer array `nums`, return an array `answer` such that `answer[i]` is equal to the product of all elements of `nums` except `nums[i]`.

The solution should not use division, and it must run in linear time for the optimal approach.

### Expected Input

- `nums`: a list of integers

### Expected Output

- A list where each position stores the product of all other elements except itself

### Solution 1: Brute Force Approach

For each index, multiply every element except the current one.

#### Algorithm

1. Create an empty result list.
2. For each index `i`, initialize a running product as `1`.
3. Loop through the array again with index `j`.
4. If `j` is not equal to `i`, multiply the running product by `nums[j]`.
5. Append the running product to the result list.
6. Return the result list after processing all indices.


```python
def product_except_self_bruteforce(nums):
    result = []

    for i in range(len(nums)):
        product = 1
        for j in range(len(nums)):
            if i != j:
                product *= nums[j]
        result.append(product)

    return result
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(n)` for the output list

### Solution 2: Optimal Approach

Build prefix products and suffix products without using division. The value at each index becomes the product of all numbers to its left multiplied by the product of all numbers to its right.

#### Algorithm

1. Create a result list filled with `1`.
2. Traverse from left to right, storing prefix products in `result[i]`.
3. Keep a variable `prefix` that represents the product of all elements before the current index.
4. Traverse from right to left with another variable `suffix`.
5. Multiply `result[i]` by `suffix` to include the product of elements to the right.
6. Update `suffix` as you move left.
7. Return the result list.


```python
def product_except_self_optimal(nums):
    result = [1] * len(nums)

    prefix = 1
    for i in range(len(nums)):
        result[i] = prefix
        prefix *= nums[i]

    suffix = 1
    for i in range(len(nums) - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)` extra space if the output array is not counted, otherwise `O(n)` including the result array


## Problem 5: Maximum Subarray

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/maximum-subarray/)

### Problem Description

Given an integer array `nums`, find the contiguous subarray with the largest sum and return that sum.

### Expected Input

- `nums`: a list of integers

### Expected Output

- An integer representing the maximum possible sum of any contiguous subarray

### Solution 1: Brute Force Approach

Enumerate every possible subarray, compute its sum, and keep the best sum seen so far.

#### Algorithm

1. Initialize `best_sum` as negative infinity.
2. Choose a starting index `i`.
3. From that start, extend the subarray to every ending index `j`.
4. Maintain a running sum while extending the end index.
5. Update `best_sum` whenever the running sum is larger.
6. Return `best_sum` after all subarrays are checked.


```python
def max_subarray_bruteforce(nums):
    best_sum = float("-inf")

    for i in range(len(nums)):
        current_sum = 0
        for j in range(i, len(nums)):
            current_sum += nums[j]
            best_sum = max(best_sum, current_sum)

    return best_sum
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use Kadane's algorithm. At each index, decide whether to extend the current subarray or start a new subarray from the current element.

#### Algorithm

1. Initialize `current_sum` and `best_sum` with the first element.
2. Traverse the array starting from the second element.
3. For each number, set `current_sum` to the larger of:
   - the current number alone
   - the current number plus the previous `current_sum`
4. Update `best_sum` if `current_sum` becomes larger.
5. Return `best_sum`.


```python
def max_subarray_optimal(nums):
    current_sum = nums[0]
    best_sum = nums[0]

    for num in nums[1:]:
        current_sum = max(num, current_sum + num)
        best_sum = max(best_sum, current_sum)

    return best_sum
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`


## Problem 6: Maximum Product Subarray

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/maximum-product-subarray/)

### Problem Description

Given an integer array `nums`, find a contiguous non-empty subarray within the array that has the largest product and return that product.

### Expected Input

- `nums`: a list of integers

### Expected Output

- An integer representing the maximum product of any contiguous subarray

### Solution 1: Brute Force Approach

Compute the product of every possible contiguous subarray and track the largest value.

#### Algorithm

1. Initialize `best_product` as negative infinity.
2. Choose a starting index `i`.
3. Extend the subarray to each ending index `j`.
4. Maintain a running product while extending.
5. Update `best_product` whenever the running product is larger.
6. Return `best_product` after checking all subarrays.


```python
def max_product_bruteforce(nums):
    best_product = float("-inf")

    for i in range(len(nums)):
        current_product = 1
        for j in range(i, len(nums)):
            current_product *= nums[j]
            best_product = max(best_product, current_product)

    return best_product
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Track both the maximum and minimum product ending at the current position, because multiplying by a negative number can swap a very small product into the new maximum product.

#### Algorithm

1. Initialize `current_max`, `current_min`, and `best_product` with the first element.
2. Traverse the rest of the array.
3. If the current number is negative, swap `current_max` and `current_min`.
4. Update `current_max` as the larger of the current number and `current_max * current number`.
5. Update `current_min` as the smaller of the current number and `current_min * current number`.
6. Update `best_product` using `current_max`.
7. Return `best_product`.


```python
def max_product_optimal(nums):
    current_max = nums[0]
    current_min = nums[0]
    best_product = nums[0]

    for num in nums[1:]:
        if num < 0:
            current_max, current_min = current_min, current_max

        current_max = max(num, current_max * num)
        current_min = min(num, current_min * num)
        best_product = max(best_product, current_max)

    return best_product
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`


## Problem 7: Find Minimum in Rotated Sorted Array

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/)

### Problem Description

Suppose an array of distinct integers was originally sorted in ascending order and then rotated some number of times. Return the minimum element in the rotated array.

### Expected Input

- `nums`: a rotated sorted list of distinct integers

### Expected Output

- The minimum integer in the array

### Solution 1: Brute Force Approach

Scan every value and keep the smallest one found.

#### Algorithm

1. Initialize `minimum` with the first element.
2. Traverse the remaining elements in the array.
3. Compare each value with `minimum`.
4. Update `minimum` whenever a smaller value is found.
5. Return `minimum` after the scan finishes.


```python
def find_min_bruteforce(nums):
    minimum = nums[0]

    for num in nums[1:]:
        minimum = min(minimum, num)

    return minimum
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use binary search. If the middle element is greater than the rightmost element, the minimum must be in the right half. Otherwise it lies in the left half including the middle element.

#### Algorithm

1. Set `left = 0` and `right = len(nums) - 1`.
2. While `left < right`, compute `mid`.
3. Compare `nums[mid]` with `nums[right]`.
4. If `nums[mid] > nums[right]`, move `left` to `mid + 1`.
5. Otherwise move `right` to `mid`.
6. When the loop ends, return `nums[left]`.


```python
def find_min_optimal(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]
```

#### Complexity

- Time Complexity: `O(log n)`
- Space Complexity: `O(1)`


## Problem 8: Search in Rotated Sorted Array

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/search-in-rotated-sorted-array/)

### Problem Description

Given a rotated sorted array of distinct integers `nums` and a target value `target`, return the index of `target` if it exists. Otherwise return `-1`.

### Expected Input

- `nums`: a rotated sorted list of distinct integers
- `target`: an integer to search for

### Expected Output

- The index of `target` if found, otherwise `-1`

### Solution 1: Brute Force Approach

Scan the array from left to right until the target is found.

#### Algorithm

1. Traverse the array using index `i`.
2. Compare `nums[i]` with `target`.
3. If they are equal, return `i`.
4. If the loop ends without finding the target, return `-1`.


```python
def search_rotated_bruteforce(nums, target):
    for i, num in enumerate(nums):
        if num == target:
            return i

    return -1
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use modified binary search. In each step, one half of the array is guaranteed to be sorted, and we can decide whether the target must lie in that half or the other half.

#### Algorithm

1. Set `left = 0` and `right = len(nums) - 1`.
2. While `left <= right`, compute `mid`.
3. If `nums[mid] == target`, return `mid`.
4. Determine whether the left half or right half is sorted.
5. Check whether the target lies inside the sorted half.
6. Narrow the search range to the half that can contain the target.
7. Return `-1` if the loop finishes without finding it.


```python
def search_rotated_optimal(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid

        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1
```

#### Complexity

- Time Complexity: `O(log n)`
- Space Complexity: `O(1)`


## Problem 9: 3Sum

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/3sum/)

### Problem Description

Given an integer array `nums`, return all unique triplets `[nums[i], nums[j], nums[k]]` such that `i`, `j`, and `k` are distinct indices and the sum of the three numbers is `0`.

The result must not contain duplicate triplets.

### Expected Input

- `nums`: a list of integers

### Expected Output

- A list of unique triplets whose sum is zero

### Solution 1: Brute Force Approach

Check every combination of three elements, collect the triplets whose sum is zero, and use a set to avoid duplicates.

#### Algorithm

1. Create an empty set to store unique triplets.
2. Use three nested loops to choose indices `i`, `j`, and `k`.
3. If `nums[i] + nums[j] + nums[k] == 0`, sort the triplet values.
4. Add the sorted triplet to the set.
5. After checking all triples, convert the set contents back to a list of lists.
6. Return the result.


```python
def three_sum_bruteforce(nums):
    triplets = set()

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            for k in range(j + 1, len(nums)):
                if nums[i] + nums[j] + nums[k] == 0:
                    triplet = tuple(sorted((nums[i], nums[j], nums[k])))
                    triplets.add(triplet)

    return [list(triplet) for triplet in triplets]
```

#### Complexity

- Time Complexity: `O(n^3)`
- Space Complexity: `O(m)` for the set of unique triplets, where `m` is the number of valid triplets

### Solution 2: Optimal Approach

Sort the array, then fix one element at a time and use two pointers to find the remaining two values. Skip duplicates so the output remains unique.

#### Algorithm

1. Sort the array.
2. Traverse the array with index `i` and treat `nums[i]` as the first element.
3. Skip `nums[i]` if it is the same as the previous value.
4. Set two pointers: `left = i + 1` and `right = len(nums) - 1`.
5. Compute the sum of the three values.
6. If the sum is too small, move `left` rightward. If too large, move `right` leftward.
7. If the sum is zero, record the triplet and move both pointers while skipping duplicates.
8. Return the list of collected triplets.


```python
def three_sum_optimal(nums):
    nums.sort()
    result = []

    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left = i + 1
        right = len(nums) - 1

        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                result.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1

                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1

    return result
```

#### Complexity

- Time Complexity: `O(n^2)` after sorting
- Space Complexity: `O(m)` for the output list, ignoring the cost of the returned answer


## Problem 10: Container With Most Water

Category: Array

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/container-with-most-water/)

### Problem Description

Given an array `height` where each value represents the height of a vertical line, find two lines that together with the x-axis form a container that holds the maximum amount of water.

Return the maximum area of water that can be contained.

### Expected Input

- `height`: a list of non-negative integers

### Expected Output

- An integer representing the maximum water area

### Solution 1: Brute Force Approach

Try every pair of lines, compute the water area for that pair, and keep the largest area.

#### Algorithm

1. Initialize `best_area` as `0`.
2. Use two nested loops to choose indices `i` and `j`.
3. Compute the width as `j - i`.
4. Compute the height of the container as `min(height[i], height[j])`.
5. Compute the area and update `best_area` if needed.
6. Return `best_area` after all pairs are checked.


```python
def max_area_bruteforce(height):
    best_area = 0

    for i in range(len(height)):
        for j in range(i + 1, len(height)):
            width = j - i
            area = width * min(height[i], height[j])
            best_area = max(best_area, area)

    return best_area
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use two pointers at the ends of the array. The area is limited by the shorter line, so move the pointer at the shorter line inward in hopes of finding a taller line that can improve the area.

#### Algorithm

1. Initialize `left` at the start and `right` at the end of the array.
2. Compute the area formed by the two lines.
3. Update `best_area` if the new area is larger.
4. Compare the two heights.
5. Move the pointer pointing to the shorter line inward.
6. Repeat until the pointers cross.
7. Return `best_area`.


```python
def max_area_optimal(height):
    left = 0
    right = len(height) - 1
    best_area = 0

    while left < right:
        width = right - left
        area = width * min(height[left], height[right])
        best_area = max(best_area, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return best_area
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`


## Problem 11: Sum of Two Integers

Category: Binary

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/sum-of-two-integers/)

### Problem Description

Given two integers `a` and `b`, return their sum without using the `+` or `-` operators.

### Expected Input

- `a`: an integer
- `b`: an integer

### Expected Output

- An integer equal to the sum of `a` and `b`

### Solution 1: Brute Force Approach

Simulate addition by repeatedly incrementing one number and decrementing the other until the second number becomes zero.

#### Algorithm

1. While `b` is not zero, adjust `a` and `b`.
2. If `b` is positive, increment `a` by `1` and decrement `b` by `1`.
3. If `b` is negative, decrement `a` by `1` and increment `b` by `1`.
4. Continue until `b` becomes zero.
5. Return `a`.


```python
def get_sum_bruteforce(a, b):
    while b != 0:
        if b > 0:
            a += 1
            b -= 1
        else:
            a -= 1
            b += 1

    return a
```

#### Complexity

- Time Complexity: `O(|b|)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use bit manipulation. XOR gives the sum without carry, and AND followed by a left shift gives the carry. Repeat until no carry remains.

#### Algorithm

1. Use a bit mask to simulate 32-bit integer behavior.
2. While `b` is not zero:
3. Compute the partial sum with `a ^ b`.
4. Compute the carry with `(a & b) << 1`.
5. Update `a` with the partial sum and `b` with the carry.
6. Convert the final 32-bit result back to a signed integer if needed.
7. Return the result.


```python
def get_sum_optimal(a, b):
    mask = 0xFFFFFFFF
    max_int = 0x7FFFFFFF

    while b != 0:
        a, b = (a ^ b) & mask, ((a & b) << 1) & mask

    return a if a <= max_int else ~(a ^ mask)
```

#### Complexity

- Time Complexity: `O(1)` for fixed-width integers
- Space Complexity: `O(1)`


## Problem 12: Number of 1 Bits

Category: Binary

Difficulty: Easy

LeetCode: [Link](https://leetcode.com/problems/number-of-1-bits/)

### Problem Description

Given a positive integer `n`, return the number of `1` bits in its binary representation.

This count is also called the Hamming weight.

### Expected Input

- `n`: a non-negative integer

### Expected Output

- An integer representing how many `1` bits appear in the binary form of `n`

### Solution 1: Brute Force Approach

Convert the integer to binary and count the number of `1` characters.

#### Algorithm

1. Convert `n` to its binary string form.
2. Initialize a counter to `0`.
3. Traverse each character in the binary string.
4. If the character is `'1'`, increment the counter.
5. Return the counter.


```python
def hamming_weight_bruteforce(n):
    count = 0

    for bit in bin(n):
        if bit == "1":
            count += 1

    return count
```

#### Complexity

- Time Complexity: `O(k)`, where `k` is the number of bits
- Space Complexity: `O(k)` because of the binary string

### Solution 2: Optimal Approach

Use Brian Kernighan's bit trick. Each time we apply `n &= n - 1`, the lowest set bit is removed.

#### Algorithm

1. Initialize `count` as `0`.
2. While `n` is not zero:
3. Replace `n` with `n & (n - 1)`.
4. Increment `count`.
5. Return `count` when `n` becomes zero.


```python
def hamming_weight_optimal(n):
    count = 0

    while n:
        n &= n - 1
        count += 1

    return count
```

#### Complexity

- Time Complexity: `O(s)`, where `s` is the number of set bits
- Space Complexity: `O(1)`


## Problem 13: Counting Bits

Category: Binary

Difficulty: Easy

LeetCode: [Link](https://leetcode.com/problems/counting-bits/)

### Problem Description

```python
def count_bits_bruteforce(n):
    result = []

    for num in range(n + 1):
        count = 0
        value = num

        while value:
            count += value & 1
            value >>= 1

        result.append(count)

    return result
```

#### Complexity

### Solution 2: Optimal Approach

#### Algorithm

3. Set `result[i] = result[i >> 1] + (i & 1)`.
4. Return the result list.


```python
def count_bits_optimal(n):
    result = [0] * (n + 1)

    for i in range(1, n + 1):
        result[i] = result[i >> 1] + (i & 1)

    return result
```

#### Complexity

- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

### Problem Description

You are climbing a staircase with `n` steps. Each time, you can climb either `1` step or `2` steps.

Return the number of distinct ways to reach the top.

### Expected Input

- `n`: a positive integer representing the number of steps

### Expected Output

- An integer representing the number of distinct ways to reach step `n`

### Solution 1: Brute Force Approach

Use recursion to try both choices at every step: climb `1` step or climb `2` steps.

#### Algorithm

1. Define a recursive function for the remaining steps.
2. If remaining steps are exactly `0`, return `1`.
3. If remaining steps are negative, return `0`.
4. Otherwise, recursively count ways from `remaining - 1` and `remaining - 2`.
5. Return the sum of those two recursive calls.


```python
def climb_stairs_bruteforce(n):
    def dfs(remaining):
        if remaining == 0:
            return 1
        if remaining < 0:
            return 0
        return dfs(remaining - 1) + dfs(remaining - 2)

    return dfs(n)
```

#### Complexity

- Time Complexity: `O(2^n)`
- Space Complexity: `O(n)` from the recursion stack

### Solution 2: Optimal Approach

This problem follows the Fibonacci pattern. The number of ways to reach step `i` equals the sum of ways to reach steps `i - 1` and `i - 2`.

#### Algorithm

1. If `n` is `1` or `2`, return `n`.
2. Track the number of ways to reach the previous two steps.
3. Iterate from step `3` to step `n`.
4. Compute the current number of ways as the sum of the previous two.
5. Shift the two tracked values forward.
6. Return the final computed value.


```python
def climb_stairs_optimal(n):
    if n <= 2:
        return n

    first = 1
    second = 2

    for _ in range(3, n + 1):
        first, second = second, first + second

    return second
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`

Category: Dynamic Programming

LeetCode: [Link](https://leetcode.com/problems/coin-change/)

### Problem Description

Given an array `coins` representing coin denominations and an integer `amount`, return the fewest number of coins needed to make up that amount.

If the amount cannot be formed, return `-1`.

### Expected Input

- `coins`: a list of positive integers
- `amount`: a non-negative integer

### Expected Output

- The minimum number of coins needed to form `amount`, or `-1` if impossible

### Solution 1: Brute Force Approach

Try every possible coin choice recursively and return the minimum number of coins among all valid combinations.

#### Algorithm

1. Define a recursive function for the remaining amount.
2. If the remaining amount is `0`, return `0`.
3. If the remaining amount is negative, return infinity to mark an invalid path.
4. For each coin, recursively solve for `remaining - coin`.
5. Add `1` coin to each valid recursive result.
6. Return the minimum result across all coins.
7. If the final answer is infinity, return `-1`.


```python
def coin_change_bruteforce(coins, amount):
    def dfs(remaining):
        if remaining == 0:
            return 0
        if remaining < 0:
            return float("inf")

        best = float("inf")
        for coin in coins:
            best = min(best, 1 + dfs(remaining - coin))

        return best

    answer = dfs(amount)
    return answer if answer != float("inf") else -1
```

#### Complexity

- Time Complexity: Exponential in the worst case because the recursion explores many repeated subproblems
- Space Complexity: `O(amount)` from the recursion depth

### Solution 2: Optimal Approach

Use bottom-up dynamic programming. Let `dp[x]` store the minimum number of coins needed to form amount `x`.

#### Algorithm

1. Create a DP array of size `amount + 1` initialized to infinity.
2. Set `dp[0] = 0` because zero coins are needed to form amount `0`.
3. For each amount from `1` to `amount`:
4. Try every coin denomination.
5. If the coin can be used, update `dp[current]` using `dp[current - coin] + 1`.
6. After filling the table, return `dp[amount]` if finite, otherwise return `-1`.


```python
def coin_change_optimal(coins, amount):
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0

    for current in range(1, amount + 1):
        for coin in coins:
            if coin <= current:
                dp[current] = min(dp[current], dp[current - coin] + 1)

    return dp[amount] if dp[amount] != float("inf") else -1
```

#### Complexity

- Time Complexity: `O(amount * len(coins))`
- Space Complexity: `O(amount)`


## Problem 18: Longest Increasing Subsequence

Category: Dynamic Programming

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/longest-increasing-subsequence/)

### Problem Description

Given an integer array `nums`, return the length of the longest strictly increasing subsequence.

A subsequence does not need to be contiguous, but it must preserve the original order.

### Expected Input

- `nums`: a list of integers

### Expected Output

- An integer representing the length of the longest increasing subsequence

### Solution 1: Brute Force Approach

Explore all subsequences recursively and keep the maximum length among those that remain strictly increasing.

#### Algorithm

1. Define a recursive function with the current index and previous chosen value.
2. If the index reaches the end of the array, return `0`.
3. Recursively compute the result of skipping the current element.
4. If the current element is greater than the previous chosen value, recursively compute the result of taking it.
5. Return the larger of the take and skip results.


```python
def lis_bruteforce(nums):
    def dfs(index, prev):
        if index == len(nums):
            return 0

        best = dfs(index + 1, prev)
        if prev is None or nums[index] > prev:
            best = max(best, 1 + dfs(index + 1, nums[index]))

        return best

    return dfs(0, None)
```

#### Complexity

- Time Complexity: Exponential because each element can be taken or skipped
- Space Complexity: `O(n)` from recursion depth

### Solution 2: Optimal Approach

Use dynamic programming where `dp[i]` stores the length of the longest increasing subsequence ending at index `i`.

#### Algorithm

1. Create a DP array of length `n` filled with `1`.
2. For each index `i`, look at all earlier indices `j`.
3. If `nums[j] < nums[i]`, then `nums[i]` can extend the subsequence ending at `j`.
4. Update `dp[i] = max(dp[i], dp[j] + 1)`.
5. Return the maximum value in `dp`.


```python
def lis_optimal(nums):
    if not nums:
        return 0

    dp = [1] * len(nums)

    for i in range(len(nums)):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(n)`


## Problem 19: Longest Common Subsequence

Category: Dynamic Programming

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/longest-common-subsequence/)

### Problem Description

Given two strings `text1` and `text2`, return the length of their longest common subsequence.

A subsequence preserves character order but does not need to be contiguous.

### Expected Input

- `text1`: a string
- `text2`: a string

### Expected Output

- An integer representing the length of the longest common subsequence

### Solution 1: Brute Force Approach

Use recursion to compare the two strings from left to right. At each mismatch, try skipping one character from either string.

#### Algorithm

1. Define a recursive function using indices `i` and `j`.
2. If either index reaches the end of its string, return `0`.
3. If `text1[i] == text2[j]`, return `1 + recurse(i + 1, j + 1)`.
4. Otherwise, recurse by skipping one character from `text1` and one from `text2`.
5. Return the larger result of those two choices.


```python
def lcs_bruteforce(text1, text2):
    def dfs(i, j):
        if i == len(text1) or j == len(text2):
            return 0
        if text1[i] == text2[j]:
            return 1 + dfs(i + 1, j + 1)
        return max(dfs(i + 1, j), dfs(i, j + 1))

    return dfs(0, 0)
```

#### Complexity

- Time Complexity: Exponential because overlapping subproblems are recomputed many times
- Space Complexity: `O(m + n)` from recursion depth

### Solution 2: Optimal Approach

Use a 2D DP table where `dp[i][j]` stores the longest common subsequence length between `text1[i:]` and `text2[j:]`.

#### Algorithm

1. Create a 2D table with dimensions `(len(text1) + 1) x (len(text2) + 1)` filled with `0`.
2. Traverse the table from bottom-right to top-left.
3. If the current characters match, set `dp[i][j] = 1 + dp[i + 1][j + 1]`.
4. Otherwise set `dp[i][j] = max(dp[i + 1][j], dp[i][j + 1])`.
5. Return `dp[0][0]`.


```python
def lcs_optimal(text1, text2):
    rows = len(text1) + 1
    cols = len(text2) + 1
    dp = [[0] * cols for _ in range(rows)]

    for i in range(len(text1) - 1, -1, -1):
        for j in range(len(text2) - 1, -1, -1):
            if text1[i] == text2[j]:
                dp[i][j] = 1 + dp[i + 1][j + 1]
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j + 1])

    return dp[0][0]
```

#### Complexity

- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`

Given a string `s` and a dictionary of strings `wordDict`, return `True` if `s` can be segmented into a space-separated sequence of one or more dictionary words.

#### Algorithm

1. Convert the dictionary to a set for fast lookup.
2. Define a recursive function starting at a given index.
3. If the index reaches the end of the string, return `True`.
4. Try every ending position for the next word.
5. If the substring is in the dictionary and the remaining suffix is breakable, return `True`.
6. If no valid split works, return `False`.


```python
def word_break_bruteforce(s, word_dict):
    words = set(word_dict)

    def dfs(start):
        if start == len(s):
            return True

        for end in range(start + 1, len(s) + 1):
            if s[start:end] in words and dfs(end):
                return True

        return False

    return dfs(0)
```

#### Complexity

### Expected Output

- An integer representing the number of ordered combinations that sum to `target`

### Solution 1: Brute Force Approach

Use recursion to try every number as the next choice until the remaining target becomes `0` or negative.

#### Algorithm

1. Define a recursive function for the remaining target.
2. If the remaining target is `0`, return `1`.

```python
def combination_sum4_bruteforce(nums, target):
    def dfs(remaining):
        if remaining == 0:
            return 1
        if remaining < 0:
            return 0

        total = 0
        for num in nums:
            total += dfs(remaining - num)

        return total

    return dfs(target)
```

#### Complexity

- Time Complexity: Exponential because many target values are recomputed repeatedly

### Solution 2: Optimal Approach

Use dynamic programming where `dp[x]` stores the number of ordered combinations that sum to `x`.

1. Create a DP array of length `target + 1` filled with `0`.

- Time Complexity: `O(target * len(nums))`

- An integer representing the maximum amount that can be robbed

At each house, choose between robbing it and skipping the next house, or skipping it and moving to the next one.

1. Define a recursive function for the current index.
2. If the index goes beyond the array, return `0`.
3. Compute the profit from robbing the current house plus the result from `index + 2`.
4. Compute the profit from skipping the current house and moving to `index + 1`.
5. Return the larger of those two choices.

#### Complexity

- Time Complexity: Exponential because the same subproblems are revisited
- Space Complexity: `O(n)` from recursion depth

### Solution 2: Optimal Approach

Use dynamic programming with two rolling values: the best result up to the previous house and the best result up to the house before that.

1. Initialize two values to represent the best result excluding and including recent houses.

5. Update the rolling values.
6. Return the final best value.


```python
def house_robber_optimal(nums):
    prev_two = 0
    prev_one = 0

    for amount in nums:
        prev_two, prev_one = prev_one, max(prev_one, prev_two + amount)

    return prev_one
```

#### Complexity

- Time Complexity: `O(n)`

Category: Dynamic Programming

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/house-robber-ii/)

### Problem Description

This is the circular version of House Robber. The first and last houses are adjacent, so they cannot both be robbed.

Return the maximum amount of money you can rob.

### Expected Input

- `nums`: a list of non-negative integers arranged in a circle

### Expected Output

- An integer representing the maximum money that can be robbed

### Solution 1: Brute Force Approach

Solve the circular restriction by exploring recursive choices while tracking whether the first house was robbed.

#### Algorithm

1. Define a recursive function with the current index and a flag indicating whether the first house was robbed.
2. If the index moves beyond the last house, return `0`.
3. If the current house is the last one and the first house was robbed, you cannot rob it.
4. At each house, compute the best result from robbing or skipping it.
5. Return the maximum result.


```python
def house_robber_ii_bruteforce(nums):
    if len(nums) == 1:
        return nums[0]

    def dfs(index, robbed_first):
        if index >= len(nums):
            return 0
        if index == len(nums) - 1 and robbed_first:
            return 0

        rob = nums[index] + dfs(index + 2, robbed_first)
        skip = dfs(index + 1, robbed_first)
        return max(rob, skip)

    return max(dfs(0, True), dfs(1, False))
```

#### Complexity

- Time Complexity: Exponential due to repeated recursive branching
- Space Complexity: `O(n)` from recursion depth

### Solution 2: Optimal Approach

Convert the circular problem into two linear House Robber problems:

- rob from house `0` to `n - 2`
- rob from house `1` to `n - 1`

Take the maximum of those two cases.

#### Algorithm

1. If there is only one house, return its value.
2. Define a helper that solves the linear House Robber problem.
3. Run the helper on `nums[:-1]`.
4. Run the helper on `nums[1:]`.
5. Return the larger result.


```python
def house_robber_ii_optimal(nums):
    if len(nums) == 1:
        return nums[0]

    def rob_linear(houses):
        prev_two = 0
        prev_one = 0

        for amount in houses:
            prev_two, prev_one = prev_one, max(prev_one, prev_two + amount)

        return prev_one

    return max(rob_linear(nums[:-1]), rob_linear(nums[1:]))
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`

### Expected Input

- `s`: a string of digits

### Expected Output

- An integer representing the number of valid decodings

### Solution 1: Brute Force Approach

Use recursion to try decoding one digit or two digits at a time whenever the chosen chunk forms a valid letter.

#### Algorithm

1. Define a recursive function starting at index `i`.

4. Recursively decode one digit by calling the function on `i + 1`.
5. If the next two digits form a number between `10` and `26`, also decode two digits by calling the function on `i + 2`.
6. Return the sum of both valid choices.


```python
def decode_ways_bruteforce(s):
    def dfs(i):
        if i == len(s):
            return 1
        if s[i] == "0":
            return 0

        total = dfs(i + 1)

        if i + 1 < len(s) and 10 <= int(s[i:i + 2]) <= 26:
            total += dfs(i + 2)

        return total

    return dfs(0)
```

#### Complexity

- Time Complexity: Exponential because the same suffixes are recomputed

5. Otherwise set `dp[i] = dp[i + 1]`.
6. If the two-digit number starting at `i` is valid, add `dp[i + 2]`.
7. Return `dp[0]`.


```python
def decode_ways_optimal(s):
    dp = [0] * (len(s) + 1)
    dp[len(s)] = 1

    for i in range(len(s) - 1, -1, -1):
        if s[i] == "0":
            dp[i] = 0
        else:
            dp[i] = dp[i + 1]
            if i + 1 < len(s) and 10 <= int(s[i:i + 2]) <= 26:
                dp[i] += dp[i + 2]

    return dp[0]
```

#### Complexity

- Time Complexity: `O(n)`

### Problem Description

You are located at the top-left corner of an `m x n` grid and want to reach the bottom-right corner.

You can only move either down or right at any step.

Return the number of unique paths.

### Expected Input

- `m`: number of rows
- `n`: number of columns

### Expected Output

- An integer representing the number of unique right-and-down paths

### Solution 1: Brute Force Approach

Use recursion to try moving right and down from each cell until the destination or an invalid cell is reached.

#### Algorithm

1. Define a recursive function for cell `(row, col)`.
2. If the cell is outside the grid, return `0`.
3. If the cell is the bottom-right corner, return `1`.
4. Recursively count paths by moving down.
5. Recursively count paths by moving right.
6. Return the sum of both counts.


```python
def unique_paths_bruteforce(m, n):
    def dfs(row, col):
        if row >= m or col >= n:
            return 0
        if row == m - 1 and col == n - 1:
            return 1
        return dfs(row + 1, col) + dfs(row, col + 1)

    return dfs(0, 0)
```

#### Complexity

- Time Complexity: Exponential because many cells are revisited through different recursive paths
- Space Complexity: `O(m + n)` from recursion depth

### Solution 2: Optimal Approach

Use dynamic programming where each cell stores the number of ways to reach the destination from that cell.

#### Algorithm

1. Create a grid filled with `1` values.
2. The last row and last column remain `1` because there is only one way from those cells.
3. Traverse the grid from bottom-right toward top-left.
4. For each cell, set its value as the sum of the cell below and the cell to the right.
5. Return the value at the top-left cell.


```python
def unique_paths_optimal(m, n):
    dp = [[1] * n for _ in range(m)]

    for row in range(m - 2, -1, -1):
        for col in range(n - 2, -1, -1):
            dp[row][col] = dp[row + 1][col] + dp[row][col + 1]

    return dp[0][0]
```

#### Complexity

- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`


## Problem 26: Jump Game

Category: Dynamic Programming

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/jump-game/)

### Problem Description

Given an array `nums` where each element represents your maximum jump length from that position, return `True` if you can reach the last index and `False` otherwise.

### Expected Input

- `nums`: a list of non-negative integers

### Expected Output

- A boolean indicating whether the last index is reachable

### Solution 1: Brute Force Approach

Use recursion to try every possible jump length from the current position until the last index is reached or all paths fail.

#### Algorithm

1. Define a recursive function for the current index.
2. If the index is at or beyond the last index, return `True`.
3. From the current index, try all jump sizes from `1` to `nums[index]`.
4. Recursively check whether any jump can reach the end.
5. If one jump succeeds, return `True`.
6. If all jumps fail, return `False`.


```python
def can_jump_bruteforce(nums):
    def dfs(index):
        if index >= len(nums) - 1:
            return True

        furthest = min(index + nums[index], len(nums) - 1)
        for next_index in range(index + 1, furthest + 1):
            if dfs(next_index):
                return True

        return False

    return dfs(0)
```

#### Complexity

- Time Complexity: Exponential in the worst case because many jump paths are explored
- Space Complexity: `O(n)` from recursion depth

### Solution 2: Optimal Approach

Use a greedy dynamic-programming-style observation: track the furthest reachable index while scanning the array.

#### Algorithm

1. Initialize `furthest` as `0`.
2. Traverse the array by index.
3. If the current index is greater than `furthest`, return `False` because it is unreachable.
4. Update `furthest` as the maximum of its current value and `index + nums[index]`.
5. If `furthest` reaches or passes the last index, return `True`.
6. If the loop finishes, return `True`.


```python
def can_jump_optimal(nums):
    furthest = 0

    for index, jump in enumerate(nums):
        if index > furthest:
            return False
        furthest = max(furthest, index + jump)
        if furthest >= len(nums) - 1:
            return True

    return True
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`


## Problem 27: Clone Graph

Category: Graph

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/clone-graph/)

### Problem Description

Given a reference to a node in a connected undirected graph, return a deep copy of the entire graph.

Each node contains a value and a list of its neighbors.

### Expected Input

- `node`: a reference to a graph node, or `None`

### Expected Output

- A new graph node representing the root of a deep-copied graph

### Solution 1: Brute Force Approach

Use DFS to visit every node and recreate it. Store a mapping from original nodes to copied nodes so cycles do not cause infinite recursion.

#### Algorithm

1. If the input node is `None`, return `None`.
2. Create a dictionary to map original nodes to cloned nodes.
3. Define a recursive DFS function.
4. If a node was already cloned, return the stored copy.
5. Create a copy of the current node.
6. Recursively clone each neighbor and append it to the copied node's neighbors.
7. Return the cloned node.


```python
class GraphNode:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []


def clone_graph_bruteforce(node):
    if not node:
        return None

    copies = {}

    def dfs(current):
        if current in copies:
            return copies[current]

        clone = GraphNode(current.val)
        copies[current] = clone

        for neighbor in current.neighbors:
            clone.neighbors.append(dfs(neighbor))

        return clone

    return dfs(node)
```

#### Complexity

- Time Complexity: `O(V + E)` because every vertex and edge is visited once
- Space Complexity: `O(V)` for the hash map and recursion stack

### Solution 2: Optimal Approach

Use BFS with a queue to clone the graph iteratively. This avoids recursion depth issues while still using a map to prevent duplicate copies.

#### Algorithm

1. If the input node is `None`, return `None`.
2. Create a clone of the starting node and store it in a dictionary.
3. Initialize a queue with the starting node.
4. While the queue is not empty, pop a node.
5. For each neighbor, create its clone if needed and add the neighbor to the copied adjacency list.
6. If a neighbor has not been processed before, push it into the queue.
7. Return the clone of the starting node.


```python
from collections import deque


def clone_graph_optimal(node):
    if not node:
        return None

    copies = {node: GraphNode(node.val)}
    queue = deque([node])

    while queue:
        current = queue.popleft()

        for neighbor in current.neighbors:
            if neighbor not in copies:
                copies[neighbor] = GraphNode(neighbor.val)
                queue.append(neighbor)

            copies[current].neighbors.append(copies[neighbor])

    return copies[node]
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V)`


## Problem 28: Course Schedule

Category: Graph

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/course-schedule/)

### Problem Description

There are `numCourses` labeled from `0` to `numCourses - 1`. Each prerequisite pair `[a, b]` means you must take course `b` before course `a`.

Return `True` if it is possible to finish all courses, otherwise return `False`.

### Expected Input

- `numCourses`: total number of courses
- `prerequisites`: a list of pairs `[course, prerequisite]`

### Expected Output

- A boolean indicating whether all courses can be completed

### Solution 1: Brute Force Approach

For each course, run DFS and track the current recursion path to detect cycles.

#### Algorithm

1. Build an adjacency list from the prerequisite pairs.
2. For each course, start a DFS if it has not already been fully processed.
3. Maintain a recursion-path set for the current DFS.
4. If DFS reaches a course already in the current path, a cycle exists.
5. If all DFS traversals complete without a cycle, return `True`.
6. Otherwise return `False`.


```python
def can_finish_bruteforce(num_courses, prerequisites):
    graph = {course: [] for course in range(num_courses)}
    for course, prereq in prerequisites:
        graph[course].append(prereq)

    visiting = set()
    visited = set()

    def dfs(course):
        if course in visiting:
            return False
        if course in visited:
            return True

        visiting.add(course)
        for prereq in graph[course]:
            if not dfs(prereq):
                return False
        visiting.remove(course)
        visited.add(course)
        return True

    for course in range(num_courses):
        if not dfs(course):
            return False

    return True
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V + E)`

### Solution 2: Optimal Approach

Use Kahn's algorithm for topological sorting. If all courses can be processed in topological order, then no cycle exists.

#### Algorithm

1. Build an adjacency list and indegree array.
2. Add all courses with indegree `0` to a queue.
3. Repeatedly remove a course from the queue.
4. Decrease the indegree of its dependent courses.
5. If a dependent course reaches indegree `0`, add it to the queue.
6. Count how many courses are processed.
7. Return whether the processed count equals `numCourses`.


```python
def can_finish_optimal(num_courses, prerequisites):
    graph = {course: [] for course in range(num_courses)}
    indegree = [0] * num_courses

    for course, prereq in prerequisites:
        graph[prereq].append(course)
        indegree[course] += 1

    queue = deque([course for course in range(num_courses) if indegree[course] == 0])
    processed = 0

    while queue:
        course = queue.popleft()
        processed += 1

        for neighbor in graph[course]:
            indegree[neighbor] -= 1
            if indegree[neighbor] == 0:
                queue.append(neighbor)

    return processed == num_courses
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V + E)`


## Problem 29: Pacific Atlantic Water Flow

Category: Graph

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/pacific-atlantic-water-flow/)

### Problem Description

You are given an `m x n` matrix of heights. Water can flow from a cell to neighboring cells with height less than or equal to the current cell.

Return all coordinates from which water can flow to both the Pacific and Atlantic oceans.

### Expected Input

- `heights`: a 2D list of integers

### Expected Output

- A list of coordinate pairs `[row, col]` that can reach both oceans

### Solution 1: Brute Force Approach

From each cell, run DFS to see whether water can reach the Pacific and Atlantic boundaries.

#### Algorithm

1. For each cell in the grid, start a DFS.
2. During DFS, move only to neighboring cells with height less than or equal to the current cell.
3. Track whether the search reaches the Pacific boundary.
4. Track whether the search reaches the Atlantic boundary.
5. If both are reachable, add the cell to the result list.
6. Repeat for all cells.


```python
def pacific_atlantic_bruteforce(heights):
    if not heights or not heights[0]:
        return []

    rows = len(heights)
    cols = len(heights[0])
    result = []
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    for row in range(rows):
        for col in range(cols):
            visited = set()

            def dfs(r, c):
                if (r, c) in visited:
                    return False, False
                visited.add((r, c))

                pacific = r == 0 or c == 0
                atlantic = r == rows - 1 or c == cols - 1

                for dr, dc in directions:
                    nr, nc = r + dr, c + dc
                    if 0 <= nr < rows and 0 <= nc < cols and heights[nr][nc] <= heights[r][c]:
                        p, a = dfs(nr, nc)
                        pacific = pacific or p
                        atlantic = atlantic or a

                return pacific, atlantic

            pacific, atlantic = dfs(row, col)
            if pacific and atlantic:
                result.append([row, col])

    return result
```

#### Complexity

- Time Complexity: `O((m * n)^2)` in the worst case because DFS is run from every cell
- Space Complexity: `O(m * n)` for the visited set and recursion stack

### Solution 2: Optimal Approach

Run reverse DFS/BFS from the ocean borders instead of from each cell. From an ocean, move inward only to cells of greater or equal height. Cells reachable from both oceans are the answer.

#### Algorithm

1. Create two visited sets: one for the Pacific and one for the Atlantic.
2. Start DFS from all Pacific-border cells and mark every cell reachable in reverse flow.
3. Start DFS from all Atlantic-border cells and mark every cell reachable in reverse flow.
4. A reverse move is valid only if the neighbor height is greater than or equal to the current height.
5. Intersect the two visited sets.
6. Return the common coordinates.


```python
def pacific_atlantic_optimal(heights):
    if not heights or not heights[0]:
        return []

    rows = len(heights)
    cols = len(heights[0])
    pacific = set()
    atlantic = set()
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    def dfs(row, col, visited):
        if (row, col) in visited:
            return
        visited.add((row, col))

        for dr, dc in directions:
            nr, nc = row + dr, col + dc
            if 0 <= nr < rows and 0 <= nc < cols and heights[nr][nc] >= heights[row][col]:
                dfs(nr, nc, visited)

    for row in range(rows):
        dfs(row, 0, pacific)
        dfs(row, cols - 1, atlantic)

    for col in range(cols):
        dfs(0, col, pacific)
        dfs(rows - 1, col, atlantic)

    result = []
    for row in range(rows):
        for col in range(cols):
            if (row, col) in pacific and (row, col) in atlantic:
                result.append([row, col])

    return result
```

#### Complexity

- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`


## Problem 30: Number of Islands

Category: Graph

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/number-of-islands/)

### Problem Description

Given a 2D grid of `'1'`s and `'0'`s, count how many islands exist. An island is formed by adjacent land cells connected horizontally or vertically.

### Expected Input

- `grid`: a 2D list of strings containing `'1'` and `'0'`

### Expected Output

- An integer representing the number of islands

### Solution 1: Brute Force Approach

For each land cell, use DFS to explore its full island while marking visited cells so the same island is not counted twice.

#### Algorithm

1. Initialize `islands` as `0`.
2. Traverse each cell of the grid.
3. When an unvisited land cell is found, increment `islands`.
4. Run DFS to mark all connected land cells as visited.
5. Continue scanning the grid.
6. Return `islands`.


```python
def num_islands_bruteforce(grid):
    if not grid or not grid[0]:
        return 0

    rows = len(grid)
    cols = len(grid[0])
    visited = set()

    def dfs(row, col):
        if (
            row < 0
            or row >= rows
            or col < 0
            or col >= cols
            or grid[row][col] == "0"
            or (row, col) in visited
        ):
            return

        visited.add((row, col))
        dfs(row + 1, col)
        dfs(row - 1, col)
        dfs(row, col + 1)
        dfs(row, col - 1)

    islands = 0
    for row in range(rows):
        for col in range(cols):
            if grid[row][col] == "1" and (row, col) not in visited:
                islands += 1
                dfs(row, col)

    return islands
```

#### Complexity

- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`

### Solution 2: Optimal Approach

Use BFS iteratively to flood-fill each island. This avoids deep recursion and still processes each land cell once.

#### Algorithm

1. Traverse the grid cell by cell.
2. When an unvisited land cell is found, increment the island count.
3. Push that cell into a queue.
4. While the queue is not empty, pop a cell and visit its valid land neighbors.
5. Mark neighbors as visited when they are queued.
6. Continue until the whole island is processed.
7. Return the number of islands.


```python
def num_islands_optimal(grid):
    if not grid or not grid[0]:
        return 0

    rows = len(grid)
    cols = len(grid[0])
    visited = set()
    islands = 0

    for row in range(rows):
        for col in range(cols):
            if grid[row][col] != "1" or (row, col) in visited:
                continue

            islands += 1
            queue = deque([(row, col)])
            visited.add((row, col))

            while queue:
                r, c = queue.popleft()

                for nr, nc in ((r + 1, c), (r - 1, c), (r, c + 1), (r, c - 1)):
                    if (
                        0 <= nr < rows
                        and 0 <= nc < cols
                        and grid[nr][nc] == "1"
                        and (nr, nc) not in visited
                    ):
                        visited.add((nr, nc))
                        queue.append((nr, nc))

    return islands
```

#### Complexity

- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`

### Problem Description

Given an unsorted array of integers `nums`, return the length of the longest consecutive elements sequence.

The sequence values do not need to be adjacent in the original array.

### Expected Input

- `nums`: a list of integers

### Expected Output

- An integer representing the length of the longest consecutive sequence

### Solution 1: Brute Force Approach

For each number, repeatedly check whether the next consecutive number exists and count the sequence length.

#### Algorithm

1. Initialize `best` as `0`.
2. For each number in `nums`, treat it as the start of a sequence.
3. Repeatedly search for `current + 1` in the array.
4. Count how long the consecutive run continues.
5. Update `best` if the current run is longer.
6. Return `best`.


```python
def longest_consecutive_bruteforce(nums):
    best = 0

    for num in nums:
        current = num
        length = 1

        while current + 1 in nums:
            current += 1
            length += 1

        best = max(best, length)

    return best
```

#### Complexity

- Time Complexity: `O(n^2)` in the worst case because list membership checks are linear
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use a set. Only start counting from numbers that do not have a predecessor. This ensures each sequence is processed once.

#### Algorithm

1. Convert `nums` to a set.
2. Initialize `best` as `0`.
3. For each number in the set, check whether `num - 1` is absent.
4. If yes, this number starts a sequence.
5. Count how long the sequence continues using `num + 1`, `num + 2`, and so on.
6. Update `best` with the maximum length found.
7. Return `best`.


```python
def longest_consecutive_optimal(nums):
    num_set = set(nums)
    best = 0

    for num in num_set:
        if num - 1 not in num_set:
            current = num
            length = 1

            while current + 1 in num_set:
                current += 1
                length += 1

            best = max(best, length)

    return best
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(n)`

### Expected Output

- A string representing one valid character order, or an empty string if no valid order exists

### Solution 1: Brute Force Approach

Build ordering constraints between characters, then use DFS-based topological sorting with cycle detection.

#### Algorithm

1. Initialize a graph containing all unique characters.
2. Compare each adjacent pair of words.
3. Find the first differing character and add a directed edge.
4. If a longer word appears before its own prefix, return an empty string.
5. Run DFS on each character with visiting and visited sets to detect cycles.
6. Append characters in postorder.
7. Reverse the collected order and return it.


```python
def alien_order_bruteforce(words):
    graph = {char: set() for word in words for char in word}

    for i in range(len(words) - 1):
        first = words[i]
        second = words[i + 1]
        min_length = min(len(first), len(second))

        if len(first) > len(second) and first[:min_length] == second[:min_length]:
            return ""

        for j in range(min_length):
            if first[j] != second[j]:
                graph[first[j]].add(second[j])
                break

    visiting = set()
    visited = set()
    order = []

    def dfs(char):
        if char in visiting:
            return False
        if char in visited:
            return True

        visiting.add(char)
        for neighbor in graph[char]:
            if not dfs(neighbor):
                return False
        visiting.remove(char)
        visited.add(char)
        order.append(char)
        return True

    for char in graph:
        if not dfs(char):
            return ""

    return "".join(reversed(order))
```

#### Complexity

- Time Complexity: `O(C)`, where `C` is the total number of characters across all words

5. Reduce the indegree of its neighbors.
6. Add any neighbor whose indegree becomes `0`.

8. Otherwise return the built string.

## Problem 33: Graph Valid Tree

Category: Graph

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/graph-valid-tree/)

### Problem Description

Given `n` nodes labeled from `0` to `n - 1` and a list of undirected edges, determine whether the graph forms a valid tree.

A valid tree must be connected and acyclic.

### Expected Input

- `n`: the number of nodes
- `edges`: a list of undirected edges

### Expected Output

- A boolean indicating whether the graph is a valid tree

### Solution 1: Brute Force Approach

Use DFS to detect cycles and confirm that all nodes are connected.

#### Algorithm

1. If the number of edges is not `n - 1`, return `False`.
2. Build an adjacency list.
3. Run DFS from node `0`, keeping track of the parent node.
4. If DFS reaches a previously visited node that is not the parent, a cycle exists.
5. After DFS, check whether all nodes were visited.
6. Return `True` only if the graph is connected and acyclic.


```python
def valid_tree_bruteforce(n, edges):
    if len(edges) != n - 1:
        return False

    graph = {node: [] for node in range(n)}
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    visited = set()

    def dfs(node, parent):
        if node in visited:
            return False

        visited.add(node)
        for neighbor in graph[node]:
            if neighbor == parent:
                continue
            if not dfs(neighbor, node):
                return False
        return True

    return dfs(0, -1) and len(visited) == n
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V + E)`

### Solution 2: Optimal Approach

Use Union-Find. A tree cannot contain a cycle, and it must have exactly `n - 1` edges.

#### Algorithm

1. If the number of edges is not `n - 1`, return `False`.
2. Initialize a Union-Find structure for all nodes.
3. For each edge, find the root of both endpoints.
4. If both endpoints already share the same root, a cycle exists, so return `False`.
5. Otherwise union the two components.
6. If all edges are processed without a cycle, return `True`.


```python
def valid_tree_optimal(n, edges):
    if len(edges) != n - 1:
        return False

    parent = list(range(n))
    rank = [1] * n

    def find(node):
        while node != parent[node]:
            parent[node] = parent[parent[node]]
            node = parent[node]
        return node

    def union(a, b):
        root_a = find(a)
        root_b = find(b)

        if root_a == root_b:
            return False

        if rank[root_a] < rank[root_b]:
            root_a, root_b = root_b, root_a

        parent[root_b] = root_a
        rank[root_a] += rank[root_b]
        return True

    for a, b in edges:
        if not union(a, b):
            return False

    return True
```

#### Complexity

- Time Complexity: `O(V + E)` with path compression and union by rank
- Space Complexity: `O(V)`

LeetCode: [Link](https://leetcode.com/problems/number-of-connected-components-in-an-undirected-graph/)

### Problem Description

Given `n` nodes labeled from `0` to `n - 1` and a list of undirected edges, return the number of connected components in the graph.

### Expected Input

- `n`: the number of nodes
- `edges`: a list of undirected edges

### Expected Output

- An integer representing the number of connected components

### Solution 1: Brute Force Approach

Use DFS from every unvisited node. Each new DFS traversal discovers one full connected component.

#### Algorithm

1. Build an adjacency list for the graph.
2. Initialize a visited set and a component counter.
3. Traverse all nodes from `0` to `n - 1`.
4. If a node is unvisited, increment the component counter.
5. Run DFS to visit every node in that component.
6. Return the component count.


```python
def count_components_bruteforce(n, edges):
    graph = {node: [] for node in range(n)}
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    visited = set()

    def dfs(node):
        if node in visited:
            return
        visited.add(node)
        for neighbor in graph[node]:
            dfs(neighbor)

    components = 0
    for node in range(n):
        if node not in visited:
            components += 1
            dfs(node)

    return components
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V + E)`

### Solution 2: Optimal Approach

Use Union-Find to merge nodes connected by edges. The number of remaining disjoint sets is the number of connected components.

#### Algorithm

1. Initialize each node as its own parent.
2. Keep a counter initialized to `n`.
3. For each edge, union its endpoints if they belong to different sets.
4. Each successful union decreases the component counter by `1`.
5. Return the final component counter.


```python
def count_components_optimal(n, edges):
    parent = list(range(n))
    rank = [1] * n
    components = n

    def find(node):
        while node != parent[node]:
            parent[node] = parent[parent[node]]
            node = parent[node]
        return node

    def union(a, b):
        nonlocal components
        root_a = find(a)
        root_b = find(b)

        if root_a == root_b:
            return

        if rank[root_a] < rank[root_b]:
            root_a, root_b = root_b, root_a

        parent[root_b] = root_a
        rank[root_a] += rank[root_b]
        components -= 1

    for a, b in edges:
        union(a, b)

    return components
```

#### Complexity

- Time Complexity: `O(V + E)`
- Space Complexity: `O(V)`

## Problem 74: Top K Frequent Elements

Category: Heap

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/top-k-frequent-elements/)

### Problem Description

Given an integer array `nums` and an integer `k`, return the `k` most frequent elements in the array.

You may return the answer in any order.

### Expected Input

- `nums`: a list of integers
- `k`: an integer such that `1 <= k <= number of unique elements`

### Expected Output

- A list containing the `k` elements that appear most frequently in `nums`

### Solution 1: Brute Force Approach

The brute force approach is to count the frequency of every unique number, sort all unique numbers by frequency in descending order, and then take the first `k` elements.

#### Algorithm

1. Create a dictionary to store the frequency of each number.
2. Traverse `nums` and update the count for each number.
3. Convert the frequency dictionary into a list of `(number, frequency)` pairs.
4. Sort this list by frequency in descending order.
5. Take the first `k` pairs from the sorted list.
6. Extract only the numbers from those pairs and return them.


```python
def top_k_frequent_bruteforce(nums, k):
    frequency = {}

    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1

    pairs = list(frequency.items())
    pairs.sort(key=lambda item: item[1], reverse=True)

    result = []
    for i in range(k):
        result.append(pairs[i][0])

    return result
```

#### Complexity

- Time Complexity: `O(n + m log m)`, where `n` is the length of `nums` and `m` is the number of unique elements. Counting takes `O(n)` and sorting takes `O(m log m)`.
- Space Complexity: `O(m)` for the frequency dictionary and sorted list.

### Solution 2: Optimal Approach

An optimal approach uses bucket sort. Since a number can appear at most `n` times, we can group numbers by their frequency and then read the buckets from high frequency to low frequency until we collect `k` elements.

#### Algorithm

1. Create a dictionary to count the frequency of each number.
2. Create a list of empty buckets where index `i` stores numbers that appear exactly `i` times.
3. Traverse the frequency dictionary and place each number into the bucket matching its frequency.
4. Start from the last bucket and move backward toward the first bucket.
5. Add numbers from each bucket into the result list.
6. Stop once the result contains `k` elements and return it.


```python
def top_k_frequent_optimal(nums, k):
    frequency = {}
    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1

    buckets = [[] for _ in range(len(nums) + 1)]
    for num, count in frequency.items():
        buckets[count].append(num)

    result = []
    for count in range(len(buckets) - 1, 0, -1):
        for num in buckets[count]:
            result.append(num)
            if len(result) == k:
                return result

    return result
```

#### Complexity

- Time Complexity: `O(n)` because counting frequencies, filling buckets, and scanning buckets together take linear time.
- Space Complexity: `O(n)` for the frequency dictionary and buckets.
