## Problem 1: Two Sum

Category: Array

Difficulty: Easy

### Problem Description

Given an integer array `nums` and an integer `target`, return the indices of the two numbers such that they add up to `target`.

You may assume exactly one valid answer exists, and the same element cannot be used twice.

### Expected Input

- `nums`: list of integers
- `target`: integer

### Expected Output

- A list of two indices

### Solution 1: Brute Force Approach

Check every possible pair until the pair sum matches the target.

#### Algorithm

1. Loop through the array with index `i`.
2. For each `i`, loop through the remaining indices `j`.
3. Check whether `nums[i] + nums[j] == target`.
4. If yes, return `[i, j]`.
5. If no pair is found, return an empty list.


```python
def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []
```

#### Complexity

- Time Complexity: `O(n^2)`
- Space Complexity: `O(1)`

### Solution 2: Optimal Approach

Use a hash map to remember previously seen values and their indices.

#### Algorithm

1. Create an empty dictionary `seen`.
2. Traverse the array once.
3. For each number, compute `target - num`.
4. If the complement is already in `seen`, return the stored index and current index.
5. Otherwise store the current number and index.
6. Return an empty list if no solution is found.


```python
def two_sum_optimal(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i

    return []
```

#### Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(n)`

## Problem 74: Top K Frequent Elements

Category: Heap

Difficulty: Medium

LeetCode: [Link](https://leetcode.com/problems/top-k-frequent-elements/)

### Problem Description

Given an integer array `nums` and an integer `k`, return the `k` most frequent elements in the array.

You may return the answer in any order.

### Expected Input

- `nums`: a list of integers
- `k`: an integer such that `1 <= k <= number of unique elements`

### Expected Output

- A list containing the `k` elements that appear most frequently in `nums`

### Solution 1: Brute Force Approach

The brute force approach is to count the frequency of every unique number, sort all unique numbers by frequency in descending order, and then take the first `k` elements.

#### Algorithm

1. Create a dictionary to store the frequency of each number.
2. Traverse `nums` and update the count for each number.
3. Convert the frequency dictionary into a list of `(number, frequency)` pairs.
4. Sort this list by frequency in descending order.
5. Take the first `k` pairs from the sorted list.
6. Extract only the numbers from those pairs and return them.


```python
def top_k_frequent_bruteforce(nums, k):
    frequency = {}

    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1

    pairs = list(frequency.items())
    pairs.sort(key=lambda item: item[1], reverse=True)

    result = []
    for i in range(k):
        result.append(pairs[i][0])

    return result
```

#### Complexity

- Time Complexity: `O(n + m log m)`, where `n` is the length of `nums` and `m` is the number of unique elements. Counting takes `O(n)` and sorting takes `O(m log m)`.
- Space Complexity: `O(m)` for the frequency dictionary and sorted list.

### Solution 2: Optimal Approach

An optimal approach uses bucket sort. Since a number can appear at most `n` times, we can group numbers by their frequency and then read the buckets from high frequency to low frequency until we collect `k` elements.

#### Algorithm

1. Create a dictionary to count the frequency of each number.
2. Create a list of empty buckets where index `i` stores numbers that appear exactly `i` times.
3. Traverse the frequency dictionary and place each number into the bucket matching its frequency.
4. Start from the last bucket and move backward toward the first bucket.
5. Add numbers from each bucket into the result list.
6. Stop once the result contains `k` elements and return it.


```python
def top_k_frequent_optimal(nums, k):
    frequency = {}
    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1

    buckets = [[] for _ in range(len(nums) + 1)]
    for num, count in frequency.items():
        buckets[count].append(num)

    result = []
    for count in range(len(buckets) - 1, 0, -1):
        for num in buckets[count]:
            result.append(num)
            if len(result) == k:
                return result

    return result
```

#### Complexity

- Time Complexity: `O(n)` because counting frequencies, filling buckets, and scanning buckets together take linear time.
- Space Complexity: `O(n)` for the frequency dictionary and buckets.
